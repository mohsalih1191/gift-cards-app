```
import requests
import threading
import json
import time

class TikTokUserInfoFetcher:
    def __init__(self):
        self.fields_mapping = {
            "Nickname": "nickname",
            "Username": "uniqueId",
            "Country": "region",
            "Language": "language",
            "About": "signature",
            "User ID": "uid",
            "SecUID": "secUid",
            "Account Created": "createTime",
            "Nickname Last Modified": "nickNameModifyTime",
            "Username Last Modified": "uniqueIdModifyTime"
        }

    def fetch_user_data(self, username):
        """
        Fetches user data from the TikTok API.

        Args:
            username (str): The TikTok username.

        Returns:
            dict: A dictionary containing the full raw user data from the API,
                  or None if an error occurred.
        """
        if not username:
            print("Error: Please provide a TikTok username.")
            return None

        url = f"https://faas-sgp1-18bc02ac.doserverless.co/api/v1/web/fn-67a396e1-78e9-4dff-8f6a-0f07c2d80c56/default/sm-t/?username={username}"
        print(f"Attempting to fetch data for @{username} from: {url}")

        try:
            response = requests.get(url, timeout=15)
            response.raise_for_status()
            data = response.json()
            print(f"Successfully fetched raw data: {json.dumps(data, indent=2)}")

            # Return the full raw data dictionary
            return data

        except requests.exceptions.ConnectionError as e:
            print(f"Connection Error: Could not connect to the API. Check your internet or API availability. Details: {e}")
            return None
        except requests.exceptions.Timeout as e:
            print(f"Timeout Error: Request to API timed out. Details: {e}")
            return None
        except requests.exceptions.HTTPError as e:
            error_message = f"HTTP Error: {e.response.status_code} - {e.response.reason}"
            try:
                error_json = e.response.json()
                if 'message' in error_json:
                    error_message += f"\nAPI Message: {error_json['message']}"
            except json.JSONDecodeError:
                pass
            print(error_message)
            return None
        except json.JSONDecodeError as e:
            print(f"JSON Decode Error: Could not parse API response. Details: {e}")
            return None
        except Exception as e:
            print(f"An unexpected error occurred: {type(e).__name__}: {e}")
            return None

# Example usage (you can uncomment and run this in a separate cell)
# fetcher = TikTokUserInfoFetcher()
# username_to_fetch = "tiktok" # Replace with the desired username
# user_info = fetcher.fetch_user_data(username_to_fetch)
#
# if user_info:
#     print("\n--- User Information ---")
#     for label, value in user_info.items():
#         print(f"{label}: {value}")
# else:
#     print("\nFailed to fetch user information.")
```

## Choose a python web framework

### Subtask:
Select a lightweight framework suitable for a simple API, such as Flask or FastAPI.


**Reasoning**:
Evaluate Flask and FastAPI for suitability and choose one based on the criteria.



```
# Evaluation of Flask vs FastAPI for a simple API fetching external data:

# Flask:
# - Ease of use: Very easy to get started, minimal boilerplate. Excellent documentation.
# - Dependencies: Lightweight, few dependencies by default.
# - Performance: Good for simple applications, but can be less performant for highly concurrent tasks compared to async frameworks.
# - Suitability for this task: Highly suitable. The task is straightforward (receive input, call a function, return output), and Flask's simplicity is a good fit.

# FastAPI:
# - Ease of use: Also easy to get started, especially with modern Python features (type hints). Automatic data validation and documentation (Swagger UI/ReDoc).
# - Dependencies: More dependencies out-of-the-box (Starlette, Pydantic, Uvicorn).
# - Performance: Excellent performance due to its asynchronous nature (ASGI).
# - Suitability for this task: Also highly suitable. While the task is simple, FastAPI's performance and automatic documentation could be beneficial if the application were to scale or handle more complex data structures. The built-in data validation using Pydantic is also a plus.

# Decision:
# For this specific task, which is a simple API fetching data from an external source, both frameworks are suitable.
# Flask is slightly simpler with fewer initial dependencies, making it a very quick option for a basic API.
# FastAPI offers better performance and automatic documentation, which are valuable features even for a simple API and would be beneficial if the project were to evolve.

# Given the goal of building a simple API and the context of potentially adding more features later, FastAPI's modern features and performance advantages make it a slightly better choice for future-proofing, even though Flask is perfectly capable of handling the current requirements.

# Therefore, choosing FastAPI.
```

```
# Create main.py
with open('main.py', 'w') as f:
    f.write("""
from fastapi import FastAPI
import requests
import json
import time

# Assume TikTokUserInfoFetcher class is available (either in this file or imported)
# For now, we'll include it here for simplicity.
class TikTokUserInfoFetcher:
    def __init__(self):
        self.fields_mapping = {
            "Nickname": "nickname",
            "Username": "uniqueId",
            "Country": "region",
            "Language": "language",
            "About": "signature",
            "User ID": "uid",
            "SecUID": "secUid",
            "Account Created": "createTime",
            "Nickname Last Modified": "nickNameModifyTime",
            "Username Last Modified": "uniqueIdModifyTime"
        }

    def fetch_user_data(self, username):
        \"\"\"
        Fetches user data from the TikTok API.

        Args:
            username (str): The TikTok username.

        Returns:
            dict: A dictionary containing the full raw user data from the API,
                  or None if an error occurred.
        \"\"\"
        if not username:
            print("Error: Please provide a TikTok username.")
            return None

        url = f"https://faas-sgp1-18bc02ac.doserverless.co/api/v1/web/fn-67a396e1-78e9-4dff-8f6a-0f07c2d80c56/default/sm-t/?username={username}"
        print(f"Attempting to fetch data for @{username} from: {url}")

        try:
            response = requests.get(url, timeout=15)
            response.raise_for_status()
            data = response.json()
            print(f"Successfully fetched raw data: {json.dumps(data, indent=2)}")

            # Return the full raw data dictionary
            return data

        except requests.exceptions.ConnectionError as e:
            print(f"Connection Error: Could not connect to the API. Check your internet or API availability. Details: {e}")
            return None
        except requests.exceptions.Timeout as e:
            print(f"Timeout Error: Request to API timed out. Details: {e}")
            return None
        except requests.exceptions.HTTPError as e:
            error_message = f"HTTP Error: {e.response.status_code} - {e.response.reason}"
            try:
                error_json = e.response.json()
                if 'message' in error_json:
                    error_message += f"\\nAPI Message: {error_json['message']}"
            except json.JSONDecodeError:
                pass
            print(error_message)
            return None
        except json.JSONDecodeError as e:
            print(f"JSON Decode Error: Could not parse API response. Details: {e}")
            return None
        except Exception as e:
            print(f"An unexpected error occurred: {type(e).__name__}: {e}")
            return None

app = FastAPI()

@app.get("/")
def read_root():
    return {"message": "TikTok User Info Fetcher API"}

@app.get("/user/{username}")
def get_user_info(username: str):
    fetcher = TikTokUserInfoFetcher()
    user_data = fetcher.fetch_user_data(username)
    if user_data:
        return user_data
    else:
        return {"error": "Could not fetch user data"}, 404 # Return 404 for not found/error

""")

# Create requirements.txt
with open('requirements.txt', 'w') as f:
    f.write("""
fastapi
requests
uvicorn
""")
```
